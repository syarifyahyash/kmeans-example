```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data-clustering.csv')
data = data.set_index('No')
data.head()
```

```python
data.info()
```

```python
data.rename(index=str, columns={'Nama':'Name'},inplace=True)
data
```

```python
data = data.drop(['Name'], axis=1)
data
```

```python
data.describe()
data.isnull().sum()
```

```python
km1 = KMeans(n_clusters=1).fit(data)

data['Labels'] = km1.labels_
plt.figure(figsize=(6,4))
sns.scatterplot(x=data['Height(cm)'], y=data['Weight(kg)'], hue=data['Labels'], palette=sns.color_palette('hls', 1))
plt.title('KMeans with 1 clusters')
```

```python
km2 = KMeans(n_clusters=2).fit(data)

data['Labels'] = km2.labels_
plt.figure(figsize=(6,4))
sns.scatterplot(x=data['Height(cm)'], y=data['Weight(kg)'], hue=data['Labels'], palette=sns.color_palette('hls', 2))
plt.title('KMeans with 2 clusters')
```

```python
km3 = KMeans(n_clusters=3).fit(data)

data['Labels'] = km3.labels_
plt.figure(figsize=(6,4))
sns.scatterplot(x=data['Height(cm)'], y=data['Weight(kg)'], hue=data['Labels'], palette=sns.color_palette('hls', 3))
plt.title('KMeans with 3 clusters')
```
