```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data-clustering.csv')
data = pd.DataFrame(data)
data = data.set_index('No')
data
```

```python
# Cluster = 3
# C1 = Basep (170, 30)
# C2 = Fasep (175, 75)
# C3 = Hasep (182, 80)

# Pilih fitur yang ingin digunakan untuk clustering
features = data[['Height(cm)', 'Weight(kg)']]

# Pilih jumlah cluster
jumlah_cluster = 3

# Inisialisasi centroid pada data 'Basep', 'Fasep', dan 'Hasep'
centroid_indices = data[data['Nama'].isin(['Basep', 'Fasep', 'Hasep'])].index
centroids = features.loc[centroid_indices]

# Fungsi untuk menghitung jarak Euclidean antara data dan centroid
def euclidean_distance(data, centroids):
    distances = []
    for i in range(jumlah_cluster):
        # Ambil centroid ke-i
        centroid_i = centroids.iloc[i]
        # Hitung jarak Euclidean dan simpan dalam list distances
        distance_i = ((data - centroid_i) ** 2).sum(axis=1) ** 0.5
        distances.append(distance_i)
    return pd.DataFrame(distances).T



# Iterasi untuk menemukan klaster
for _ in range(10):  # Jumlah iterasi, sesuaikan sesuai kebutuhan
    # Hitung jarak Euclidean
    distances = euclidean_distance(features, centroids)

    # Tambahkan kolom jarak ke DataFrame
    for i in range(jumlah_cluster):
        data[f'Jarak ke C{i + 1}'] = distances[i]

    # Tentukan klaster untuk setiap data
    data['Cluster'] = distances.idxmin(axis=1) + 1

    # Perbarui lokasi centroid
    centroids = data.groupby('Cluster')[['Height(cm)', 'Weight(kg)']].mean()

# Tampilkan hasil
data
```
